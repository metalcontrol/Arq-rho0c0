```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros
c0 = 343
f0 = 1000
N = 11
s = np.array([(n**2) % N for n in range(N)])

# Calcular la matriz de profundidades
depth_matrix = np.zeros((N, N))
for y in range(N):
    for x in range(N):
        s_sum = s[x] + s[y]
        depth_matrix[y, x] = (s_sum * c0) / (2 * N * f0)  # en metros

# Crear grfico
fig, ax = plt.subplots(figsize=(8, 8))
im = ax.imshow(depth_matrix, cmap="viridis")

# Agregar valores numricos sin "cm", centrados, fuente ms grande
for i in range(N):
    for j in range(N):
        value = f"{depth_matrix[i, j]*1000:.0f}"
        ax.text(j, i, value, ha="center", va="center", color="white", fontsize=15)

# Etiquetas de ejes
ax.set_xticks(np.arange(N))
ax.set_yticks(np.arange(N))
ax.set_xticklabels([f"m={i+1}" for i in range(N)])
ax.set_yticklabels([f"n={i+1}" for i in range(N)])

# Limpiar ttulo y etiquetas extra
plt.tight_layout()
plt.show()
```
